```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# features vs target
features = [col for col in mushrooms_ds.columns if col!='class']
mushrooms_ds_without_class = mushrooms_ds.drop(columns=['class'])
# need to get dummies
get_dummy = pd.get_dummies(mushrooms_ds_without_class)

x = get_dummy # features
y = mushrooms_ds['class'] # target

# splitting into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1, test_size=0.25)
```

```python
tree_mushrooms = DecisionTreeClassifier(random_state=1, max_depth=7)
tree_mushrooms.fit(x_train, y_train)
```

```python
print(tree_mushrooms.score(x_test, y_test))
```
